```python
def process_interconnectors(csv_path):
    print(csv_path)
    d = pd.read_csv(csv_path, index_col=0)
    columns = ['SETTLEMENTDATE', 'MWFLOW', 'INTERCONNECTORID']

    d = d.loc[d['INTERCONNECTORID'] == 'V-SA', columns]
    d.loc[:, 'SETTLEMENTDATE'] = pd.to_datetime(d.loc[:, 'SETTLEMENTDATE'])
    d.loc[:, 'start-of-interval'] = d.loc[:, 'SETTLEMENTDATE'] - pd.Timedelta('5min')
    return d

inters = [process_interconnectors(p) for p in paths]
```

```python
data = pd.concat(inters, axis=0)
data = data.sort_values('SETTLEMENTDATE')

data.to_csv('interconnectors.csv')
```
